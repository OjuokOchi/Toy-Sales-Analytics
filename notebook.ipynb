The datasets main_data and product_master have one common column **ProductCode** hence we will merge the two on the ProductCode column.

```python
#merging the datasets on product code
merged_df = pd.merge(main_data, product_master, on='ProductCode', how='inner')
merged_df.head()
merged_df.shape
```

```python
merged_df['Territory'].value_counts()
```

```python
# rows where territory is missing
missing_territory = merged_df.loc[merged_df['Territory'].isnull()]
missing_territory
```

```python
merged_df['Country'].value_counts()
```

```python
#Filling rows where territory is missing and country is USA with NA(North America)
merged_df.loc[(merged_df['Territory'].isnull()) & (merged_df['Country'] =='USA'), 'Territory'] = 'NA'
merged_df.loc[merged_df['Country'] == 'USA'].head(10)
```

```python
merged_df.isnull().sum()/len(merged_df) * 100
```

After filling the missing values in the territory column where country is USA with NA(North America) the % of missing values in the territory column reduced to 2%

```python
missing_territory_2 = merged_df.loc[merged_df['Territory'].isnull()]
missing_territory_2
```

since Canada is also in North America we will fill territory where country is canada with NA

```python
#Filling rows where territory is missing and country is Canada with NA(North America)
merged_df.loc[(merged_df['Territory'].isnull()) & (merged_df['Country'] == 'Canada'), 'Territory'] = 'NA'
merged_df.loc[merged_df['Country'] == 'Canada'].head(10)
```
