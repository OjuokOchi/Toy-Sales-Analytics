```python
#mean and median for sales data
mean_sales = main_data['Sales'].mean()
median_sales = main_data['Sales'].median()

#plotting 
plt.figure(figsize=(10, 6))
sns.histplot(main_data['Sales'], kde=True, color='skyblue', bins=30)
plt.axvline(mean_sales, color='red', linestyle='--', label=f'Mean: {mean_sales:.2f}')
plt.axvline(median_sales, color='green', linestyle='-', label=f'Median: {median_sales:.2f}')
plt.legend()
plt.title('Distribution of Sales with Mean and Median')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
#checking null percentage wise
main_data.isnull().sum() / len(main_data) * 100
```

```python
#value counts for Territory
main_data['Territory'].value_counts()
```

```python
#filter for rows where 'Country' is 'USA' and 'Territory' is not missing
usa_with_territory = main_data[(main_data['Country'] == 'USA') & (main_data['Territory'].notnull())]

#check the number of such entries
print(f"Number of 'USA' entries with a 'Territory': {len(usa_with_territory)}")

#display the entries
usa_with_territory.head()
```

No Territory data for 'USA' country

```python
#Replace missing 'Territory' values with 'USA' where 'Country' is 'USA'
main_data.loc[(main_data['Territory'].isnull()) & (main_data['Country'] == 'USA'), 'Territory'] = 'USA'
```

Created a new teritory called USA, for all entries in the country USA that had missing values in the Territory

```python
#value counts for Territory
main_data['Territory'].value_counts()
```

```python
#null values
main_data.isnull().sum()
```

```python
#filling in 'Canada' to Territory where country was 'Canada' 
main_data.loc[(main_data['Territory'].isnull()) & (main_data['Country'] == 'Canada'), 'Territory'] = 'Canada'
```

```python
#null inspection
main_data.isnull().sum()
```

```python
main_data['Territory'].value_counts()
```

We now have Canada and USA as Territories as well!
